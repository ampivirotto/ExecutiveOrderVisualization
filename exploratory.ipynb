```python
import pandas as pd
```

```python
df = pd.read_csv('executive_orders_updated.csv')
```

```python
df = df.drop(labels = 'disposition_notes', axis = 1)
```

```python
df = df.sort_values(by=['signing_date'])
```

```python
df_filtered = df[df['signing_date'] > '1981-1-20']
```

```python
## remove commas in columns 
## remove new line characters in columns

for col in df_filtered.columns:
    df_filtered[col] = df_filtered[col].astype(str).str.replace(',', '')
    df_filtered[col] = df_filtered[col].str.replace('\n', ' ')

```

```python
df_filtered.to_csv('executive_orders_updated.csv', index=False)
```
